```python
import numpy as np
import pandas as pd
import pylab as pl
import os

%pylab inline

#poisson, gaussian, chisquare, binomial, gamma, lognormal
plt.style.use('ggplot')

```

```python
distributions = {}
distributions['gaussian'] = {}
distributions['poisson'] = {}
distributions['chisquare'] = {}
distributions['binomial'] = {}
distributions['gamma'] = {}
distributions['lognormal'] = {}

pop_mean = 100 
pop_sd = 100
```

```python
np.random.seed(277) 
dist_sizes = np.random.choice(np.arange(10,2000), size = 100, replace = False)
#print(dist_sizes)
```

```python
for size in dist_sizes:
    distributions['gaussian'].update({
            size : np.random.normal(pop_mean,pop_sd,size)
        })
    distributions['poisson'].update({
            size : np.random.poisson(pop_mean,size)
        })
    distributions['chisquare'].update({
            size : np.random.chisquare(pop_mean,size)
        })
    distributions['binomial'].update({#To Do
            size : np.random.normal(pop_mean,pop_sd,size)
        })
    distributions['gamma'].update({#To DO
            size : np.random.normal(pop_mean,pop_sd,size)
        })
    distributions['lognormal'].update({#To DO
            size : np.random.normal(pop_mean,pop_sd,size)
        })

#print distributions['gamma']
```
